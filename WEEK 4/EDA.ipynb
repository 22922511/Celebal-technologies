# **Q  Bivariate Analysis**

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

#**bivariate numerical - categorical**

```python
df=sns.load_dataset("titanic")
```

```python
df.head()
```

```python
sns.barplot(x="class",y="fare",hue="sex",data=df)
```

```python
sns.barplot(y="sex",x="fare",data=df)
```

```python
sns.boxplot(x="class",y="age",data=df)
```

```python
sns.violinplot(x="class",y="age",data=df)
```

```python
x=df[df["class"]=="First"]
sns.distplot(x["age"])
```

# **Bivariate analysis categorical**

```python
df.head()
```

```python
sns.barplot(y="survived",hue="class",data=df)
```

```python
sns.countplot(x="survived",hue="class",data=df)
```

```python
x=pd.crosstab(df["survived"],df["class"])
x
```

```python
sns.heatmap( x,annot=True,fmt=".1f",cmap="coolwarm")
```
